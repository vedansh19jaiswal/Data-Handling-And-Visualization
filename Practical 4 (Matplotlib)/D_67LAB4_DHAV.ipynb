```python
x=np.array(sales['month_number'])
y=np.array(sales['total_units'])
plt.plot(x, y, linestyle = 'dotted' ,linewidth='3', marker = 'o', color ='red' )
plt.title('EACH MONTH PROFIT')
plt.ylabel("profits")
plt.legend(["blue"], loc="lower right")
plt.xlabel("month")
```

```python
p1 = sales['facecream']
p2 = sales['facewash']
p3 = sales['toothpaste']
p4 = sales['bathingsoap']
p5 = sales['shampoo']
p6 = sales['moisturizer']
x = sales['month_number']

plt.plot(x,p1,marker='o',color='red' )
plt.plot(x,p2,marker='o',color='yellow' )
plt.plot(x,p3,marker='o',color='orange' )
plt.plot(x,p4,marker='o',color='blue' )
plt.plot(x,p5,marker='o',color='green' )
plt.plot(x,p6,marker='o' )
plt.xlabel=('Months')
plt.ylabel('Products')
plt.legend(["Facecream","Facewash","ToothPaste","BathingSoap","Shampoo","Moisturizer"],loc='center right')
```
