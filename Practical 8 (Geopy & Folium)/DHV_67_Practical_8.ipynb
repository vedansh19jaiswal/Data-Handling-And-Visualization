Task 1: Consider Names of two places from the user and find out the distance between them using
Geopy library and Mark the location using folium.

```python
from geopy.distance import geodesic as GD
```

```python
loc1 = input("Enter Location1 : ")
loc2 = input("Enter Location2 : ")
world=folium.Map()
geolocator = Nominatim(user_agent="my_raequest")
location1 = geolocator.geocode(loc1)
location2 = geolocator.geocode(loc2)
la1 , lo1 = location1.latitude, location1.longitude;
la2 , lo2 = location2.latitude, location2.longitude;
points = []
place_lat=[la1,la2]
place_lng=[lo1,lo2]
for i in range(len(place_lat)):
    points.append([place_lat[i], place_lng[i]])

city1 = (la1,lo1)
city2 = (la2,lo2)
for index,lat in enumerate(place_lat):
    folium.Marker([lat,
                   place_lng[index]],).add_to(world)
folium.PolyLine(points, color='red').add_to(world)
#folium.Marker(
 #    location=[location1.latitude,location1.longitude,location2.latitude,location2.longitude],
  #   folium.PolyLine(points, color='red')
   #  ).add_to(world)

print("The distance between "+loc1+" and "+loc2+" is: ", GD(city1,city2).km)
world
```

```python
import pandas as pd

data=['Ramdeobaba College of Engineering','Savita Apartments','Lata Mangeshkar Hospital', 'Orange Hospital','Hotel Ashoka', 'Eternity Mall Nagpur', 'Deekshabhoomi','Tekdi Ganesh Mandir']

df = pd.DataFrame(data, columns=['Names'])
df
```

```python
from geopy.geocoders import Nominatim

df[['lat', 'long']] = df['Names'].apply(
    geolocator.geocode).apply(lambda x: pd.Series(
        [x.latitude, x.longitude], index=['lat', 'long']))
df
```

```python
def select_marker_icon(df):
    if df['Names'] == 'Savita Apartments':
        return 'home'
    elif df['Names'] == 'Lata Mangeshkar Hospital':
        return 'plus'
    elif df['Names'] == 'Kingsway Hospital':
        return 'plus'
    elif df['Names'] == 'Hotel Ashoka':
        return 'hotel'
    return 'info-sign'
```

```python
df['Icon'] = df.apply(select_marker_icon, axis=1)
df
```

```python
world = folium.Map()

for _, city in df.iterrows():
    folium.Marker(
        location=[city['lat'], city['long']],
        popup=city['Names'],
        tooltip=[city['Names']],
        icon=folium.Icon(icon=city['Icon'],prefix= 'fa')
    ).add_to(world)


world
```
