Consider the text file with following URL &amp; perform the following operation using Regular Expression

url = &#39;https://www.gutenberg.org/files/2638/2638-0.txt&#39;

```python
import re
import requests
the_idiot_url = 'https://www.gutenberg.org/files/2638/2638-0.txt'

def get_book(url):
    # Sends a http request to get the text from project Gutenberg
    raw = requests.get(url).text
    # Discards the metadata from the beginning of the book
    start = re.search(r"\*\*\* START OF THIS PROJECT GUTENBERG EBOOK .* \*\*\*",raw )
    # Discards the text starting Part 2 of the book
    stop = re.search(r"subscribe", raw).start()
    # Keeps the relevant text
    text = raw[start:stop]
    return text

def preprocess(sentence):
    return re.sub('[^A-Za-z0-9.]+' , ' ', sentence).lower()

book = get_book(the_idiot_url)
processed_book = preprocess(book)
print(processed_book)
```

Consider the following Phone numbers:


*   123-122-78999

The general format of a phone number is as follows:

*   Starts with 3 digits and ‘-‘ sign

iii) Password Verification:
Write a Python program to check the validity of a password using Regular expression.
Validation Rules:
* At least 1 letter between [a-z A-Z].
* At least 1 number between [0-9].
* At least 1 character from [&amp;#@].
* Minimum length 6 characters.

```python
pwd=input("Enter the password you want to validate ")
match=re.search(r'^(?=.*[A-Za-z])(?=.*\d)(?=.*[&#@])[A-Za-z\d&#@]{8,}$', pwd)

if match:
 print("Valid Password")

else:
  print("Password not valid")
```

```python
import re
p= input("Input your password:- ")
x = True
while x:
    if (len(p)<6 or len(p)>12):
        break
    elif not re.search("[a-z]",p):
        break
    elif not re.search("[0-9]",p):
        break
    elif not re.search("[A-Z]",p):
        break
    elif not re.search("[$#@]",p):
        break
    elif re.search("\s",p):
        break
    else:
        print("Valid Password")
        x=False
        break

if x:
    print("Not a Valid Password")
```

1. Problem Statement – Scrapping all of the phone numbers from a website for a
requirement by making use of Python Regular Expressions &amp; save it in CSV/ list
Website URL: http://www.summet.com/dmsi/html/codesamples/addresses.html

```python
import urllib.request
import re
url = "https://www.summet.com/dmsi/html/codesamples/addresses.html"


response = urllib.request.urlopen(url)

html = response.read()

htmlStr = html.decode()

pdata = re.findall("\(\d{3}\) \d{3}-\d{4}", htmlStr)

for item in pdata:
    print(item)
```
