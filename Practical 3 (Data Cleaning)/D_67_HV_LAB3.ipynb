4)Replace the NaN in Age column by median value.


```python
Toyota_dataset['Age'].fillna(Toyota_dataset['Age'].median(), inplace=True)
display(Toyota_dataset)
```

5) Replace the string value in column Door with number.

```python
Toyota_dataset.Doors[Toyota_dataset.Doors == 'one'] = 1
Toyota_dataset.Doors[Toyota_dataset.Doors == 'two'] = 2
Toyota_dataset.Doors[Toyota_dataset.Doors == 'three'] = 3
Toyota_dataset.Doors[Toyota_dataset.Doors == 'four'] = 4
Toyota_dataset.Doors[Toyota_dataset.Doors == 'five'] = 5
Toyota_dataset.Doors[Toyota_dataset.Doors == 'six'] = 6
display(Toyota_dataset)
```

6)Returns a Series containing counts of unique values for column ‘FuelType’ (Petrol, Diesel, CNG )

7)Get the mode value of FuelType and  fill NA/NaN values using the specified value of mode.

```python
Toyota_dataset['FuelType'].fillna(Toyota_dataset['FuelType'].mode()[0], inplace=True)
display(Toyota_dataset)
```

8) Get the mode value of Metcolor and fill NA/NaN values using the specified value of mode.

```python
Toyota_dataset['MetColor'].fillna(Toyota_dataset['MetColor'].mode()[0], inplace=True)
display(Toyota_dataset)
```

9) Replace the value of FuelType as for Petrol replaces with 0, diesel with 1 and CNG with 2.

```python
Toyota_dataset.FuelType[Toyota_dataset.FuelType == 'Petrol'] = 0
Toyota_dataset.FuelType[Toyota_dataset.FuelType == 'Diesel'] = 1
Toyota_dataset.FuelType[Toyota_dataset.FuelType == 'CNG'] = 2
display(Toyota_dataset)
```
